```python
import pandas as pd
df = pd.read_csv('trs_repos.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df['created_at']=pd.to_datetime(df['created_at']).dt.strftime('%Y-%m-%d')
df.drop_duplicates()
df
```

```python
df['language_id'] = df['language'].astype('category').cat.codes
```

```python
df.sort_values(by=['language_id'])
dfl = df[['language_id', 'language']]
dfl = df.drop_duplicates(subset=['language_id'])
dfl.sort_values(by=['language_id'])
```

```python
df['owner_id'] = df['name'].astype('category').cat.codes
```

```python
df.sort_values(by=['owner_id'])
dfo = df[['owner_id', 'name']]
dfo = df.drop_duplicates(subset=['owner_id'])
dfo.sort_values(by=['owner_id'])
```

```python
dc = pd.read_csv('contributors.csv')
dc.dropna()
dc.drop_duplicates()
dc
```

```python
dc['contributor_id'] = dc['username'].astype('category').cat.codes
```

```python
dc.sort_values(by=['contributor_id'])
dc = dc[['contributor_id', 'username', 'contributions', 'repository_url']]
dc = dc.drop_duplicates(subset=['contributor_id'])
dc.sort_values(by=['contributor_id'])
```

```python
import pyodbc

connection = pyodbc.connect('DRIVER={SQL SERVER};SERVER=LAPTOP-JESJKN69\SQLEXPRESS;DATABASE=github_db')
cursor = connection.cursor()
```

```python


# Define SQL statements
sql_statements = [
    '''
    SET XACT_ABORT ON;
    BEGIN TRANSACTION QUICKDBD;
    ''',

    '''
    USE github_db;
    CREATE TABLE [Repositories] (
        [name] VARCHAR(255) NOT NULL,
        [url] VARCHAR(255) NOT NULL,
        [description] VARCHAR(MAX),
        [stars] INT NOT NULL,
        [created_at] DATETIME NOT NULL,
        [language_id] INT NOT NULL,
        [forks] INT NOT NULL,
        [open_issues] INT NOT NULL,
        [owner_id] INT NOT NULL,
        CONSTRAINT [PK_Repositories] PRIMARY KEY CLUSTERED ([url] ASC)
    );
    ''',

    '''
    CREATE TABLE [Language] (
        [language_id] INT NOT NULL,
        [name] VARCHAR(255) NOT NULL,
        CONSTRAINT [PK_Language] PRIMARY KEY CLUSTERED ([language_id] ASC)
    );
    ''',

    '''
    CREATE TABLE [Contributor] (
        [contributor_id] INT NOT NULL,
        [name] VARCHAR(255) NOT NULL,
        CONSTRAINT [PK_Contributor] PRIMARY KEY CLUSTERED ([contributor_id] ASC)
    );
    ''',

    '''
    CREATE TABLE [Owner] (
        [owner_id] INT NOT NULL,
        [name] VARCHAR(255) NOT NULL,
        CONSTRAINT [PK_Owner] PRIMARY KEY CLUSTERED ([owner_id] ASC)
    );
    ''',

    '''
    CREATE TABLE [repo_contri] (
        [contributor_id] INT NOT NULL,
        [url] VARCHAR(255) NOT NULL,
        [contributions_count] INT NOT NULL,
        CONSTRAINT [FK_repo_contri_contributor_id] FOREIGN KEY ([contributor_id]) REFERENCES [Contributor] ([contributor_id]),
        CONSTRAINT [FK_repo_contri_url] FOREIGN KEY ([url]) REFERENCES [Repositories] ([url])
    );
    ''',

    '''
    ALTER TABLE [Repositories] WITH CHECK ADD CONSTRAINT [FK_Repositories_language_id] FOREIGN KEY ([language_id]) REFERENCES [Language] ([language_id]);
    ''',

    '''
    ALTER TABLE [Repositories] WITH CHECK ADD CONSTRAINT [FK_Owner_owner_id] FOREIGN KEY ([owner_id]) REFERENCES [Owner] ([owner_id]);
    ''',

    '''
    COMMIT TRANSACTION QUICKDBD;
    '''
]

# Execute each SQL statement
for statement in sql_statements:
    cursor.execute(statement)

# Commit the transaction
connection.commit()

```

```python

# Iterate over each row in the DataFrame and insert data into the 'repo_contri' table
for row in dc.itertuples():
    # Check if the repository_url already exists in the 'Repositories' table
    cursor.execute('SELECT COUNT(*) FROM Repositories WHERE url = ?', row.repository_url)
    result = cursor.fetchone()
    if result[0] > 0:
        # If the repository_url exists, insert the data into 'repo_contri'
        cursor.execute('''
        INSERT INTO repo_contri (
           contributor_id,
           url,
           contributions_count
        )
        VALUES (?,?,?)
        ''', (row.contributor_id, row.repository_url, row.contributions))

# Commit the transaction
connection.commit()
```
